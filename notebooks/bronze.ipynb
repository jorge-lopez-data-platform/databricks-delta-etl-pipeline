```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Bronze Layer").getOrCreate()
```

```python
#read csv
df = spark.read.csv("/Workspace/Users/lopezechagarayjorge@gmail.com/databricks-delta-etl-pipeline/src/Superstore.csv", header=True, inferSchema=True)

#show schema
df.printSchema()

```

```python
#save to Unity Catalog table
df.write.mode("overwrite").option("delta.columnMapping.mode", "name").saveAsTable("bronze_sales")

print("table created at bronze layer")
```
